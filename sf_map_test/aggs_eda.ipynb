```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('nhood_aggs.csv')
```

```python
df.head()
```

```python
df['a_mean'] = df['a_predicti'] / df['PNTCNT']
df['c_mean'] = df['c_predicti'] / df['PNTCNT']
df['break_mean'] = df['break_even'] / df['PNTCNT']
```

```python
plt.figure(figsize=(14,10))
x_bar = np.arange(len(df.neighborho))
plt.bar(x_bar, df.break_mean)
plt.xticks(x_bar, df.neighborho, rotation="vertical")
plt.show()
```

```python
df2 = pd.read_csv('points_by_hood_raw.csv')
```

```python
df2.head()
```

```python
by_hood = df2.groupby('neighborho').median()
```

```python
by_hood = by_hood[['c_predicti', 'a_predicti', 'delta_at_7', 'break_even']]
```

```python
by_hood
```

```python
plt.figure(figsize=(14,10))
x_bar2 = np.arange(len(by_hood))
plt.bar(x_bar2, by_hood.break_even)
plt.xticks(x_bar2, by_hood.index, rotation="vertical")
plt.show()
```
